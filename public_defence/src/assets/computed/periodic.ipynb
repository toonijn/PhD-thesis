```python
from pyslise import PyslisePeriodic
from math import pi

def V(x):
    return -(x*(pi-x))**3

pp = PyslisePeriodic(V, 0, pi, tolerance=1e-12)
pp
```

```python
import matplotlib.pyplot as plt
import numpy as np
import json

xs = np.linspace(0, pi, 200)

data = []

for i, E, fs in pp.eigenpairsByIndex(0, 10):
    plt.title(E)
    dfs = []
    for f in fs:
        plt.plot(xs, f(xs)[0, :])
        dfs.append(list(f(xs)[0, :]))
    data.append({'E': E, 'fs': dfs})
    plt.show()

with open("periodic.json", 'w') as file:
    json.dump({'x': list(xs), 'eigs': data}, file)
```
