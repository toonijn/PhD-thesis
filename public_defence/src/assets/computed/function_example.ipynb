```python
import matplotlib.pyplot as plt
import numpy as np

ugent_blauw = "#1e64c8"

xs = np.linspace(-1, 1.5, 200)
ys = xs ** 2

fig, ax = plt.subplots(1, 1, figsize=(2,2), dpi=200)

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.plot(xs, ys, zorder=10, color=ugent_blauw)

ax.set_xticks([])
ax.set_yticks([])


ax.plot((1), (0), ls="", marker=">", ms=3, color="k",
        transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot((0), (1), ls="", marker="^", ms=3, color="k",
        transform=ax.get_xaxis_transform(), clip_on=False)

plt.savefig("function_example.svg", bbox_inches="tight")
```
