```python
import sys
sys.path.append('..')
import ugent

import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
n = 100

multiplicities = {}
for i, l in enumerate(sorted([i*i + j*j for i in range(1, n + 1) for j in range(1, n + 1)])):
    if l not in multiplicities:
        multiplicities[l] = []
    multiplicities[l].append(i)
```

```python
l = 0

for v, indexes in sorted(multiplicities.items(), key=lambda v: (len(v[1]), v[0])):
    if len(indexes) > l:
        print(len(indexes), v, indexes)
    l = len(indexes)
```

```python
import numpy as np
from math import pi
from matplotlib.colors import LightSource

i = 2
j = 3

xs = np.linspace(0, pi, 30)
X, Y = np.meshgrid(xs, xs)

f = np.sin(i * X) * np.sin(j * Y)

fs1 = 12
fs2 = 15

fig, ax = plt.subplots(subplot_kw=dict(projection='3d'), figsize=(3.5, 3.5))

ls = LightSource()
# To use a custom hillshading mode, override the built-in shading and pass
# in the rgb colors of the shaded surface calculated from "shade".
rgb = ls.shade(f, cmap=ugent.RdBu, blend_mode='soft')
surf = ax.plot_surface(X, Y, -f, rstride=1, cstride=1, facecolors=rgb,
                       linewidth=0, antialiased=False, shade=False)
ax.set_xticks([0, pi])
ax.set_xticklabels(["$0$", "$\\pi$"], fontsize=fs1)
ax.set_yticks([0, pi])
ax.set_yticklabels(["$0$", "$\\pi$"], fontsize=fs1)
ax.invert_yaxis()
ax.set_zticks([0])
ax.set_zticklabels(["$0$"], fontsize=fs1)
ax.tick_params(axis='both', which='major', pad=-2)
ax.view_init(elev=20., azim=-20)

plt.savefig("example_zero_3d.png", bbox_inches="tight", dpi=800)
plt.show()



xs = np.linspace(0, pi, 200)
X, Y = np.meshgrid(xs, xs)

f = np.sin(i * X) * np.sin(j * Y)

fig, ax = plt.subplots(figsize=(3, 3))

ls = LightSource()
# To use a custom hillshading mode, override the built-in shading and pass
# in the rgb colors of the shaded surface calculated from "shade".
surf = ax.pcolormesh(X, Y, f.T, cmap=ugent.RdBu, shading="gouraud")
ax.set_xticks([0, pi])
ax.set_xticklabels(["$0$", "$\\pi$"], fontsize=fs2)
ax.set_yticks([0, pi])
ax.set_yticklabels(["$0$", "$\\pi$"], fontsize=fs2)
ax.set_aspect(1)

plt.savefig("example_zero_2d.png", bbox_inches="tight", dpi=800)
plt.show()

```
