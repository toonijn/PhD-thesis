```python
import sys
sys.path.append('..')
import ugent

import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
n = 100

multiplicities = {}
for i, l in enumerate(sorted([i*i + j*j for i in range(1, n + 1) for j in range(1, n + 1)])):
    if l not in multiplicities:
        multiplicities[l] = []
    multiplicities[l].append(i)
```

```python
l = 0

for v, indexes in sorted(multiplicities.items(), key=lambda v: (len(v[1]), v[0])):
    if len(indexes) > l:
        print(len(indexes), v, indexes)
    l = len(indexes)
```

```python
import numpy as np
from math import pi
from matplotlib.colors import LightSource

i = 3
j = 2

xs = np.linspace(0, pi, 30)
X, Y = np.meshgrid(xs, xs)

f = np.sin(i * X) * np.sin(j * Y)


fig = plt.figure(figsize=(7,4), dpi=300)
gs = fig.add_gridspec(1, 2, width_ratios=(2.5, 4))
ax3 = fig.add_subplot(gs[0, 1], projection='3d')

ugent.style_3d(ax3)

surf = ax3.plot_surface(X, Y, -f, cmap=ugent.RdBu.reversed(), vmin=-1, vmax=1, alpha=1, linewidth=0, antialiased=True)
ax3.set_xticks([0, pi])
ax3.set_xticklabels(["$0$", "$\\pi$"])
ax3.set_yticks([0, pi])
ax3.set_yticklabels(["$0$", "$\\pi$"])
ax3.set_zticks([0])
ax3.set_zticklabels(["$0$"])
ax3.tick_params(axis='both', which='major', pad=-2)
ax3.view_init(elev=20., azim=-100)
ax3.set_box_aspect((1, 1, 0.6))

xs = np.linspace(0, pi, 200)
X, Y = np.meshgrid(xs, xs)

f = np.sin(i * X) * np.sin(j * Y)

ax = fig.add_subplot(gs[0,0])

surf = ax.pcolormesh(X, Y, f, cmap=ugent.RdBu, shading="gouraud")
ax.set_xticks([0, pi])
ax.set_xticklabels(["$0$", "$\\pi$"])
ax.set_yticks([0, pi])
ax.set_yticklabels(["$0$", "$\\pi$"])
ax.set_aspect(1)

bbox = fig.bbox_inches.from_bounds(0.6, 0.7, 5.5, 2.5)
plt.savefig("example_zero_both.png", bbox_inches=bbox, dpi=500)
plt.show()
```
