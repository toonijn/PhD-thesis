```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(1, 1, figsize=(6,2), dpi=200, sharey=True)
axes = [ax]

x = np.linspace(0, 2, 200)
f1 = .5*(x -  2.2) ** 2 - 1
f2 = np.sqrt(.8*(2.1-x)) - .8

def diff(a, h):
    r = np.ndarray(a.shape, dtype=a.dtype)
    r[1:-1] = (a[2:] - a[:-2]) / (2*h)
    r[0] = r[1]
    r[-1] = r[-2]
    return r

newton = np.abs(f1 / diff(f1, h)) < np.abs((f2 / diff(f2, h)))
not_newton = np.logical_not(newton)
smallest = np.abs(f1) < np.abs(f2)
not_smallest = np.logical_not(smallest)

for sel, ax in zip([newton], axes):
    ax.set_yticks([0])
    ax.set_xticks([])
    ax.grid(axis='y')
    
    l = 0
    while l < len(sel):
        for i in range(l, len(sel)):
            if sel[i] != sel[l]:
                break
        else:
            i = len(sel)
        a1, a2 = 1, 0.5
        ls1, ls2 = '-', ':'
        if not sel[l]:
            a1, a2 = a2, a1
            ls1, ls2 = ls2, ls1
        
        ax.plot(x[l:i], f1[l:i], color=ugent.blauw, alpha=a1, ls=ls1)
        ax.plot(x[l:i], f2[l:i], color=ugent.rood, alpha=a2, ls=ls2)

        l = i
        
plt.savefig('select_lambda_newton.pdf', bbox_inches="tight")
```
