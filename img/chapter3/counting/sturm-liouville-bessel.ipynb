```python
from pyslise import SturmLiouville

slp = SturmLiouville(lambda x: x, lambda x: .25/x, lambda x: x, 1e-6, 1, tolerance=1e-14)
```

```python
import matplotlib.pyplot as plt
import numpy as np

xs = np.linspace(0, 1, 201)
eigs = slp.eigenpairsByIndex(0, 5, (0, 1), (0, 1))

for i, E, f in eigs:
    plt.plot(xs, f(xs)[0, :])
    plt.title(f"$E_{{{i}}} = {E:.4f}$")
    plt.show()
```

```python
import pandas as pd

xs = np.linspace(0, 1, 501)
i, E, f = eigs[4]

df = pd.DataFrame(zip(xs, f(xs)[0, :]), columns=["x", "y"])
df.to_csv("sturm-liouville-data.csv")
```
