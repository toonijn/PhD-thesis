```python
import sys
sys.path.append('..')
import ugent
import matplotlib.pyplot as plt
import string
import numpy as np
import matplotlib as mpl
```

```python
mpl.rc('font', family='serif', size=10, serif="Computer Modern Roman")
mpl.rc('text', usetex=True)

x = np.linspace(-1, 1, 211)
y = np.linspace(-1, 1, 211)
X, Y = np.meshgrid(x, y)

F = X*X + Y*Y - 1
G = X*X*Y*Y

fig, axes = plt.subplots(
    ncols=2, nrows=1, figsize=(7, 2),
    dpi=300,
    sharex=True, sharey=True)
plt.subplots_adjust(right=0.8, top=1.1)

vmin = -1
vmax = 1

for i, data in enumerate([F, G]):
    ax = axes[i]
    ax.set_xticks(ticks=[-1,0,1])
    ax.set_yticks(ticks=[-1,0,1])
    ax.set_aspect(1)
    ax.set_title("$x^2 + y^2 - 1$" if i == 0 else "$x^2 y^2$", fontsize=11)
    pcolormesh = ax.pcolormesh(x, y, data.T, shading='gouraud', vmin=vmin, vmax=vmax, cmap=ugent.RdBu)

fig.colorbar(pcolormesh, ax=axes)
plt.savefig("nm_interpolation.png", bbox_inches='tight', dpi=500)
```
