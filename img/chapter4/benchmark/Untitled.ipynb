```python
data = {}
with open("output.txt") as output:
    for l in output:
        cells = l.split("\t")
        if len(cells) == 3:
            name, count, time = cells
            if name not in data:
                data[name] = []
            data[name].append((int(count), float(time[:-3])))
```
