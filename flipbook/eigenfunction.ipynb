```python
from strands import Schrodinger2D, Rectangle

s = Schrodinger2D(lambda x, y: 10*(x*x + y*y), Rectangle(-1,1,-1,1), maxBasisSize=32, gridSize=(48,48))
```

```python
eigs = s.eigenfunctions(10)
eigs
```

```python
import numpy as np
import matplotlib.pyplot as plt

E4, f4 = eigs[4]
E5, f5 = eigs[5]
print(E4, E5)
xs = np.linspace(-1,1, 1000)
X, Y = np.meshgrid(xs, xs)
Z = f4(X, Y) - 0.4*f5(X, Y)

plt.pcolormesh(X, Y, Z)
```

```python
import imageio

imageio.imwrite('eigenfunction.exr', Z.astype("float32"))
```
