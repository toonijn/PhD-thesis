```python
from strands import Schrodinger2D, Rectangle, Circle
import matplotlib.pyplot as plt
import numpy as np
from math import cos, pi
plt.rcParams['figure.dpi'] = 100

b = 5.5
V = lambda x, y: (1+x**2)*(1+y**2)

ixaru = Schrodinger2D(V, Rectangle(-b,b, -b,b,), gridSize=(48, 48), maxBasisSize=20)
ixaru.eigenvalues(10)
```

```python
xs  = 0.6*np.linspace(-b, b, 200)
ys  = 0.6*np.linspace(-b, b, 200)
X, Y = np.meshgrid(xs, ys)

for i, (E, f) in enumerate(ixaru.eigenfunctions(4)):
    Z = f(X, Y)
    vmax = np.max(np.abs(Z))
    plt.pcolormesh(X, Y, Z, cmap="RdBu", vmin=-vmax, vmax=vmax, shading="gouraud")
    plt.colorbar()
    plt.title(f"E_{i} = {E:.6f}")
    plt.show()
```
