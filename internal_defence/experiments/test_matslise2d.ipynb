```python
from pyslise2d import Pyslise2D
import matplotlib.pyplot as plt
import numpy as np
from math import cos, pi
plt.rcParams['figure.dpi'] = 100

b = 5.5
V = lambda x, y: (1+x**2)*(1+y**2)

ixaru = Pyslise2D(V, -b,b, -b,b, tolerance=1e-10)
ixaru.eigenvaluesByIndex(0, 10)
```

```python
xs  = 0.6*np.linspace(-b, b, 200)
ys  = 0.6*np.linspace(-b, b, 200)
X, Y = np.meshgrid(xs, ys)

for i, E, m in ixaru.eigenvaluesByIndex(0, 4):
    for f in ixaru.eigenfunction(E):
        Z = f(xs, ys)
        vmax = np.max(np.abs(Z))
        plt.pcolormesh(X, Y, Z, cmap="RdBu", vmin=-vmax, vmax=vmax, shading="gouraud")
        plt.colorbar()
        plt.title(f"E_{i} = {E:.6f}")
        plt.show()
```
