```python
from pyslise import Pyslise, PyslisePeriodic
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append('../../img')
import ugent
```

```python
x = np.linspace(0, 1, 1000)

n = 4

ref = -x

def curve(s):
    global ref
    def V1(x):
        return -x*x*(1-x)

    def V(x):
        return 11000*(s*V1(x**7) + V1((1-x)**12)) + 500*(x*(1-x))**6
    
    
    p = PyslisePeriodic(V, 0, 1, tolerance=1e-14)
    # i, E, f = p.eigenpairsByIndex(n, n+1, (1, 1), (-0.1, 1))[0]
    i, E, fs = p.eigenpairsByIndex(n, n+1)[0]
    f = fs[0]
    y = f(x)[0, :]
    # y /= np.max(np.abs(y))**(0.4)
    if np.count_nonzero(y * ref > 0) > np.count_nonzero(y * ref < 0):
        return -y
    return y

ref = curve(1.7)

fig, ax = plt.subplots(1, 1, figsize=(8,2))
for s in np.linspace(.6, 2.43, 20):
    ax.plot(x, -curve(s), c=ugent.blauw, lw=0.5)
ax.set_frame_on(False)
ax.set_xticks([])
ax.set_yticks([])

plt.savefig(f"invite_background.pdf", bbox_inches="tight")
plt.show()
```
